## 4.1 Little Red Riding Hood Network

Train a neural network to solve the  Little Red Riding Hood problem in sklearn and Keras. Try the neural network with different inputs and report the results.

## Solution

First of all, the arrays of the input and output data on which the neural networks will be trained are created as numpy arrays. In this case the trainind input data consists of 4-features correspoing to the contact subject characteristics and 3-samples. At the sime time, the output tensor has 3-samples and 4 categorical dimensions corresponding to the LRRH posible set of non-disjoint actions.

````python
import numpy as np
x_train=np.array([[1,1,0,0],[0,1,1,0],[0,0,0,1]])
y_train=np.array([[1,0,0,0],[0,0,1,1],[0,1,1,0]])
````

````python
import pandas as pd
df = pd.DataFrame(x_train,columns=['Big Ears', 'Big Teeth', 'Handsome', 'Wrinkled'])
print('TRAINING DATASET EXAMPLES')
df.join(pd.DataFrame(y_train,columns=['Scream', 'Hug', 'Food', 'Kiss']))
````

## LRRH Network Architecture

Both the SciKitLearn and Keras-TensorFlow neural networks will share a common structure according to the proposed design of the Little Red Riding Hood architeture, corresponding to a dense neural network with 1 hidden layer of 3 neurons, 4 inputs and 4 outputs, and a Sigmoid (Logistic) activation function on each layer.

### Sci-Kit Learn Neural Network

For starters the Sci-Kit Learn neural network classifier is trained. The multi-layer perceptron classifier is used to be able to specify the training parameters as well as the size of the hidden layers of the network. The maximum epochs of training are set to 1000, yet the early stopping condition is left untouched with default parameters.

````python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(solver='lbfgs', alpha=0.01, hidden_layer_sizes=(5, 2), random_state=1)
clf.fit(x_train,y_train)
clf.predict(x_train)
````

````python
from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(solver='lbfgs', alpha=0.01, hidden_layer_sizes=(2), random_state=1)
clf.fit(x_train,y_train)
clf.predict(x_train)
````

````python
from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(solver='lbfgs', alpha=0.01, hidden_layer_sizes=(3), random_state=1)
clf.fit(x_train,y_train)
clf.predict(x_train)
````

````python
from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(solver='lbfgs', alpha=0.01, hidden_layer_sizes=(3,2), random_state=1)
clf.fit(x_train,y_train)
clf.predict(x_train)
````

````python
from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(solver='lbfgs', alpha=0.1, hidden_layer_sizes=(3), random_state=1)
clf.fit(x_train,y_train)
clf.predict(x_train)
````

````python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(activation='logistic',solver='sgd',learning_rate_init=0.05,hidden_layer_sizes=(3),max_iter=1000,random_state=1,verbose=True)
````

The training process is stopped on the 633 epoch due to the early stippoing conditions. At the end of the process the Cross-Entropy loss funtion has a value of 0.059.

````python
clf.fit(x_train,y_train)
````

To check on the correct specification of the network architecture, the final coefficients and intercepts tensors are retrieved. The results confirm the classification algorithm correspond to a NN((4),3,4) with a total of 31 parameters, meaning the scructure can be correctly verificated.

````python
clf.coefs_
````

````python
clf.intercepts_
````

Finally, the estimations over the actions are computed using the calibrated NN. The results are present on the following table where it can be easily validated that the model has indeed learned the existing relation between the set of characteristics and actions of LRRH.

````python
df = pd.DataFrame(x_train,columns=['Big Ears', 'Big Teeth', 'Handsome', 'Wrinkled'])
print('PREDICTIONS WITH SCIKIT-LEARN NETWORK')
df.join(pd.DataFrame(clf.predict_proba(x_train),columns=['Scream', 'Hug', 'Food', 'Kiss']))
````

### Keras-TensorFlow Neural Network

Having completed the Si-Kit Learn framework NN construction and calaibration, it's posible to move forward and replicate the process using the Keras - TensorFlow backend framework.

As well as in the previous process, the modelling begins by specifing the network scructure including its size, layers, connection architecture, and activation functions.

````python
from keras.models import Sequential
from keras.layers import Dense, Activation
np.random.seed(123)

model = Sequential()

model.add(Dense(units=100, input_dim=4))
model.add(Activation('relu'))
model.add(Dense(units=4))
model.add(Activation('sigmoid'))

from keras.optimizers import SGD
model.compile(loss = 'binary_crossentropy',
              optimizer = SGD(lr=0.01, momentum=0.9, decay=0.0, nesterov=False),
              metrics = ['accuracy'])

model.fit(x_train, y_train, epochs=100, verbose=2)

print(model.predict(x_train))
````

````python
from keras.models import Sequential
from keras.layers import Dense, Activation
np.random.seed(123)

model = Sequential()

model.add(Dense(units=3, input_dim=4))
model.add(Activation('sigmoid'))
model.add(Dense(units=4))
model.add(Activation('sigmoid'))
````

Retrieving the summary of the model structure it can be verificated the the proposed architecture has been completely achieved.

````python
model.summary()
````

To have comparable results with the Sci-Kit Learn framework, the optimizations procedure, parameters, adn loss function are carried on with no modifications.

````python
from keras.optimizers import SGD
model.compile(loss = 'binary_crossentropy',
              optimizer = SGD(lr=0.05, momentum=0.9, decay=0.0, nesterov=True),
              metrics = ['accuracy'])
````

Completing the training the process at the maximum of 1000 epochs, the calibrated neural network ends with a Cross-Entropy loss function of 0.043 compared with the 0.059 achived by the Sci-Kit learn tool. This can be only attributable to the early stopping conditions as the baseline objective value is reached in the 803 epoch and keeps improving through the remaining trainig epochs.

````python
model.fit(x_train, y_train, epochs=1000, verbose=2)
````

Yet the small difference in the objective loss funciotion value, the predicted results are comparable to those of the Si-Kit Learn framework.

````python
df = pd.DataFrame(x_train,columns=['Big Ears', 'Big Teeth', 'Handsome', 'Wrinkled'])
print('PREDICTIONS WITH KERAS-TENSORFLOW NETWORK')
df.join(pd.DataFrame(model.predict(x_train),columns=['Scream', 'Hug', 'Food', 'Kiss']))
````

## 4.2 Boston House Price Prediction

In the next questions we are going to work using the dataset *Boston*. This dataset measures the influence of socioeconomical factors on the price of several estates of the city of Boston. This dataset has 506 instances, each one characterized by 13 features:

* CRIM - per capita crime rate by town
* ZN - proportion of residential land zoned for lots over 25,000 sq.ft.
* INDUS - proportion of non-retail business acres per town.
* CHAS - Charles River dummy variable (1 if tract bounds river; 0 otherwise)
* NOX - nitric oxides concentration (parts per 10 million)
* RM - average number of rooms per dwelling
* AGE - proportion of owner-occupied units built prior to 1940
* DIS - weighted distances to five Boston employment centres
* RAD - index of accessibility to radial highways
* TAX - full-value property-tax rate per 10,000 USD
* PTRATIO - pupil-teacher ratio by town
* B - $1000(Bk - 0.63)^2$ where $Bk$ is the proportion of blacks by town
* LSTAT - % lower status of the population

Output variable:
* MEDV - Median value of owner-occupied homes in 1000's USD

**Note:** In this exercise we are going to predict the price of each estate, which is represented in the `MEDV` variable. It is important to remember that we are always aiming to predict `MEDV`, no matter which explanatory variables we are using. That means, in some cases we will use a subset of the 13 previously mentioned variables, while in other cases we will use all the 13 variables. But in no case we will change the dependent variable $y$.

## 4.4 Modeling linear and non linear relationships

* Generate two new subsets filtering these characteristics:
     * $D_1$:  $X = \textit{'RM'}$, $y = \textit{'MEDV'}$
     * $D_2$:  $X = \textit{'LSTAT'}$, $y = \textit{'MEDV'}$
* For each subset, generate a training partition and a test partition using a ratio of $ 70 \% - 30 \% $

````python
D1X=boston[['RM']].values
D1Y=boston[['MEDV']].values
D2X=boston[['LSTAT']].values
D2Y=boston[['MEDV']].values
from sklearn.model_selection import train_test_split
D1X_train,D1X_test,D1Y_train,D1Y_test=train_test_split(D1X,D1Y,test_size=0.3,random_state=123)
D2X_train,D2X_test,D2Y_train,D2Y_test=train_test_split(D2X,D2Y,test_size=0.3,random_state=123)
````

* Train a linear regression model on both subsets of data:
     * Report the mean square error on the test set
     * Print the values of $ w $ and $ w_0 $ of the regression equation
     * Generate a graph where you visualize the line obtained by the regression model in conjunction with the training data and the test data
* How does the model perform on $ D_1 $ and $ D_2 $? Why?
    
    - **Answer:** The models perform fairly good considering each one of them only includes one regressor. In spite of this behaviour, graphically it can be confirmed that the linear hypothesis is too restrictive and the performance can be improved considering other funtional bases of the independent variables.

````python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
D1=LinearRegression(fit_intercept=True)
D1.fit(D1X_train.reshape(-1,1),D1Y_train)
print('----D1 LINEAR REGRESSION MODEL----')
print("Mean squared error: %.2f" % mean_squared_error(D1Y_test, D1.predict(D1X_test)))
print("W: ", D1.coef_)
print("W0: ",D1.intercept_)
plt.scatter(D1X_train, D1Y_train,  color='blue')
plt.scatter(D1X_test, D1Y_test,  color='red')
plt.plot(D1X_train, D1.predict(D1X_train), color='black', linewidth=3)
plt.legend(('Fitted','Train','Test'))
plt.ylabel('MEDV')
plt.xlabel('RM')
plt.show()
````

````python
D2=LinearRegression(fit_intercept=True)
D2.fit(D2X_train.reshape(-1,1),D2Y_train)
print('----D2 LINEAR REGRESSION MODEL----')
print("Mean squared error: %.2f" % mean_squared_error(D2Y_test, D2.predict(D2X_test)))
print("W: ", D2.coef_)
print("W0: ",D2.intercept_)
plt.scatter(D2X_train, D2Y_train,  color='blue')
plt.scatter(D2X_test, D2Y_test,  color='red')
plt.plot(D2X_train, D2.predict(D2X_train), color='black', linewidth=3)
plt.legend(('Fitted','Train','Test'))
plt.ylabel('LSTAT')
plt.xlabel('RM')
plt.show()
````

## 4.5 Training a regression model

* Generate a 70-30 partitioning of the data **using all the features**. (Do not include the dependent variable `MEDV`)
* Train a linear regression model with the objective of predicting the output variable `MEDV`.
     * Report the mean square error on the test set

````python
X=boston.loc[:,boston.columns!='MEDV']
Y=boston[['MEDV']].values
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=123)
LinReg=LinearRegression(fit_intercept=True)
LinReg.fit(X_train,Y_train)
print("Mean squared error: %.2f" % mean_squared_error(Y_test, LinReg.predict(X_test)))
````

* Train a regression model using `MLPRegressor` in order to predict the output variable` MEDV`.
     * Report the mean square error on the test set

````python
from sklearn.neural_network import MLPRegressor
MSE=[]
import warnings
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    for i in range(200):
        MLP=MLPRegressor(random_state=123,hidden_layer_sizes=(i+1),max_iter=500)
        MLP.fit(X_train,np.ravel(Y_train))
        MSE.append(mean_squared_error(Y_test, MLP.predict(X_test)))
plt.plot(MSE)
plt.xlabel('Neurons in hidden layer')
plt.ylabel('MSE on test')
plt.axvline(np.array(MSE).argmin(),color='red')
plt.show()
MLP=MLPRegressor(random_state=123,hidden_layer_sizes=(np.array(MSE).argmin()),max_iter=500)
MLP.fit(X_train,np.ravel(Y_train))
print('Best number of neurons: ',np.array(MSE).argmin())
print("Mean squared error of best model: %.2f" % mean_squared_error(Y_test, MLP.predict(X_test)))
````
